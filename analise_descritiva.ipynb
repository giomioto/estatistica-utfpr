```r
df <- iris
```

```r
summary(df)
```

```r
moda <- function(x) {
  uniq_x <- unique(x)
  uniq_x[which.max(tabulate(match(x, uniq_x)))]
}
```

```r
for (coluna in colnames(df)[1:(ncol(df) - 1)]) {
  cat("--------------------------------------------------\n")
  cat("Coluna:", coluna, "\n")
  cat("Moda", ":", moda(df[[coluna]]), "\n")
  cat("Média", ":", mean(df[[coluna]]), "\n")
  cat("Mediana", ":", median(df[[coluna]]), "\n")
  cat("Desvio", ":", sd(df[[coluna]]), "\n")
}
```

```r
par(mfrow = c(2, 2))
options(repr.plot.width = 12, repr.plot.height = 12)

boxplot(Sepal.Width ~ Species, data = df,
        col = "#015a35",
        main = "Boxplot do Largura da Sépala por Espécie",
        xlab = "Espécies",
        ylab = "Largura da Sépala")


boxplot(Sepal.Length ~ Species, data = df,
        col = "#05055f",
        main = "Boxplot do Comprimento da Sépala por Espécie",
        xlab = "Espécies",
        ylab = "Comprimento da Sépala")
boxplot(Petal.Width ~ Species, data = df,
        col = "#a50303",
        main = "Boxplot do Largura da Pétala por Espécie",
        xlab = "Espécies",
        ylab = "Largura da Pétala")


boxplot(Petal.Length ~ Species, data = df,
        col = "#c2b504",
        main = "Boxplot do Comprimento da Pétala por Espécie",
        xlab = "Espécies",
        ylab = "Comprimento da Pétala")

par(mfrow = c(1, 1))
```

```r
# # Definindo o dispositivo gráfico para salvar a imagem
# png("grafico_histogramas.png", width = 800, height = 800)

par(mfrow = c(2, 2))
options(repr.plot.width = 8, repr.plot.height = 8)

hist(df$Sepal.Length,
     col = "#8f0957",
     main = "Histograma do comprimento da sépala",
     xlab = "Comprimento da sépala",
     freq = FALSE)
x <- seq(min(df$Sepal.Length), max(df$Sepal.Length), length = 100)
lines(x, dnorm(x, mean(df$Sepal.Length), sd(df$Sepal.Length)),
      col = "blue", lwd = 2)

hist(df$Sepal.Width,
     col = "#6d0404dc",
     main = "Histograma da largura da sépala",
     xlab = "Largura da sépala",
     freq = FALSE)
x <- seq(min(df$Sepal.Width), max(df$Sepal.Width), length = 100)
lines(x, dnorm(x, mean(df$Sepal.Width), sd(df$Sepal.Width)),
      col = "blue", lwd = 2)

hist(df$Petal.Length,
     col = "#015a35",
     main = "Histograma do comprimento da pétala",
     xlab = "Comprimento da pétala",
     freq = FALSE)
x <- seq(min(df$Petal.Length), max(df$Petal.Length), length = 100)
lines(x, dnorm(x, mean(df$Petal.Length), sd(df$Petal.Length)),
      col = "blue", lwd = 2)

hist(df$Petal.Width,
     col = "#6d3706",
     main = "Histograma da largura da pétala",
     xlab = "Largura da pétala",
     freq = FALSE)
x <- seq(min(df$Petal.Width), max(df$Petal.Width), length = 100)
lines(x, dnorm(x, mean(df$Petal.Width), sd(df$Petal.Width)),
      col = "blue", lwd = 2)

par(mfrow = c(1, 1))
```
